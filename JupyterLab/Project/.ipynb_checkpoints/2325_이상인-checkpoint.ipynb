```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
idols = pd.read_csv('./kpop_idols.csv')
idols.head()
```

```python
idols.shape
```

```python
idols.info()
```

```python
idols.columns = ['StageN', 'FullN', 'KFullN', 'KStageN', 'Birthday', 'Group', 'Country', 'Birthplace', 'OtherGroup', 'Gender']
idols.head()
```

```python
# 생일을 datetime으로 변환
idols.Birthday = pd.to_datetime(idols.Birthday)
idols.info()
```

```python
# 본명 결측값 삭제
idols.drop(idols[idols.KFullN.isna()].index, inplace=True)
idols.reset_index(inplace=True, drop=True)
idols.head()
```

```python
idols.info()
```

```python
indexes=[]
for i in range(len(idols)):
    row = idols.loc[i]
    if str(row.KStageN) in str(row.KFullN):
        indexes.append(0)
    else:
        indexes.append(1)
idols['isPseudo'] = indexes
idols.head()
```

```python
idols.info()
```

## 첫 번째 가설 : 외국인이 한국인보다 가명을 많이 사용할 것이다.

```python
pseudoRates = pd.DataFrame(idols.groupby('Country').isPseudo.mean()).sort_values(by='isPseudo', ascending=False)
pseudoRates.reset_index(inplace=True)
pseudoRates
```

```python
plt.xticks(rotation=45)
sns.barplot(x='Country', y='isPseudo', data=pseudoRates)
plt.show()
```

```python
Krate = idols[idols.Country == 'South Korea'].isPseudo.mean()
Krate
```

```python
Frate = idols[idols.Country != 'South Korea'].isPseudo.mean()
Frate
```

* 외국인이 한국인보다 가명을 많이 사용할 것이라는 가설은 옳다.

## 두 번째 가설 : 더 늦게 태어난 사람들이 더 많이 가명을 사용할 것이다.

```python
idols.sort_values(by='Birthday')
```

```python
pseudoRates=[]
for i in range(1976,2006,5): # 70's ~ 00's
    tstart = pd.to_datetime(f'{i}-01-01')
    tend = pd.to_datetime(f'{i+10}-01-01')
    pseudoRates.append(idols.loc[(idols.Birthday>=tstart)&(idols.Birthday<tend),'isPseudo'].mean())
pseudoRates
```

```python
sns.lineplot(data=pseudoRates)
plt.show()
```

* 더 늦게 태어난 사람들이 더 가명을 많이 사용할 것이라는 가설은 틀렸다.

## 세 번째 가설 : 한쪽 성별이 더 가명을 많이 사용하지는 않을 것이다.

```python
pseudoRates = pd.DataFrame(idols.groupby('Gender').isPseudo.mean())
pseudoRates.reset_index(inplace=True)
pseudoRates
```

```python
sns.barplot(x='Gender', y='isPseudo', data=pseudoRates)
plt.show()
```

```python
idols.Gender.value_counts()
```

* 남성이 더 가명을 많이 사용했다. 가설은 틀렸다.

## 추가 : 상관관계 분석

```python
categoricals = idols.drop(['StageN', 'FullN', 'KFullN', 'KStageN', 'Birthday', 'Group', 'Birthplace', 'OtherGroup'], axis=1)
categoricals = pd.get_dummies(categoricals)
categoricals
```

```python
plt.figure(figsize=(12, 10))
plt.rcParams['axes.unicode_minus'] = False
sns.heatmap(data=categoricals.corr(), annot=True, cmap='Blues')
plt.show()
```

* 특별한 항목은 발견할 수 없었다.
