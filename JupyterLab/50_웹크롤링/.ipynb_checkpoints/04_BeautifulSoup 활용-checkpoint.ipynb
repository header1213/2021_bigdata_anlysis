# 4. BeautifulSoup 활용
* 멜론사이트 크롤링하기
* 노래제목, 가수명, 앨범명, 좋아요개수 추출

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.melon.com/chart/index.htm'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.229 Whale/2.10.123.42 Safari/537.36"
}
res = requests.get(url, headers = headers)
res
```

```python
res.text[:300]
```

```python
soup = BeautifulSoup(res.text)
```

## 노래 제목 추출

```python
songs = soup.select('.ellipsis.rank01')
songs
```

```python
title = []
for song in songs:
    title.append(song.find('a').get_text())
title
```

## 가수 추출

```python
songs = soup.select('.ellipsis.rank02')
singer = []
for song in songs:
    singer.append(song.find('a').get_text())
singer
```

## 앨범 추출

```python
songs = soup.select('.ellipsis.rank03')
album = []
for song in songs:
    album.append(song.find('a').get_text())
album
```

## 좋아요 추출

```python
# songs = soup.select('.lst50') + soup.select('.lst100')
songs = soup.select('.button_etc.like')
# contsids = []
params = ''
for song in songs:
    # contsids.append(song['data-song-no'])
    params += song['data-song-no'] + ','
params = params[:-1]
params
```

```python
# url = 'https://www.melon.com/commonlike/getSongLike.json?contsIds=' + '%2C'.join(contsids)
url = 'https://www.melon.com/commonlike/getSongLike.json?contsIds=' + params

res = requests.get(url, headers=headers)
res
```

```python
res.json()
```

```python
songs = res.json()['contsLike']
like = []
for song in songs:
    like.append(song['SUMMCNT'])
like
```

```python
# 제목, 가수, 앨범, 좋아요로 dataframe 생성
df_data = {
    'title': title,
    'singer': singer,
    'album': album,
    'like': like
}
df = pd.DataFrame(df_data)
df
```
