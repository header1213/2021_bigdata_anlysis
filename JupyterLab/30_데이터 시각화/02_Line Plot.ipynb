# 2. Line Plot (선 그래프)
* 시간에 따른 변화의 추이를 시각화하는 데에 유용함
* 예) 주식 변화, CO2 농도, 기온 변화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2-1. 그래프 그리기

```python
# (1, 3), (8, 10)을 연결하는 선 그리기
x = [1, 8]
y = [3, 10]
plt.plot(x, y)
plt.show()
```

```python
# 두 개의 점만 보여주기
plt.plot(x, y, 'o')
plt.show()
```

```python
# 점과 선을 둘 다 그리기
plt.plot(x, y, marker='o')
plt.show()
```

```python
# (1, 2) ~ (10, 20)
x = np.arange(1, 11)  # 1 ~ 10
y = x * 2
plt.plot(x, y, marker='o')
plt.show()
```

## 2-2. Marker, Line, Color
* Marker : https://matplotlib.org/stable/api/markers_api.html?highlight=marker
* Line : https://matplotlib.org/3.0.3/gallery/lines_bars_and_markers/line_styles_reference.html
* Color : https://matplotlib.org/stable/gallery/color/named_colors.html


```python
# Marker
plt.plot(x, y, marker='*', markersize=10)
plt.plot(y, x, marker='$hey$', ms=20)
plt.show()
```

```python
# Line Style
plt.plot(x, y, linestyle='--')
plt.plot(y, x, ls=':')
plt.show()
```

```python
# Color
plt.plot(x, y, color='b', linewidth=5)
plt.plot(y, x, c='r', lw=3)
plt.show()
```

```python
# Format String
plt.plot(x, y, 'o:r')
plt.plot(y, x, '*-.b')
plt.show()
```

```python
# 그래프 분리하기
plt.plot(x, y, 'o:b')

plt.figure()  # 새로운 Canvas를 생성
plt.plot(y, x, '*--r')

plt.show()
```

```python
# 그래프 사이즈 변경하기
plt.figure(figsize=(5, 5))
plt.plot(x, y, 'o:r')

plt.figure(figsize=(10, 5))
plt.plot(y, x, '*--b')

plt.show()
```

## 2-3. Title, Label, Tick, Legend

```python
# 타이틀
plt.title('타이틀', fontsize=20)
plt.plot(x, y)
plt.show()
```

```python
# Label 설정
plt.plot(x, y)
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize=15)
plt.show()
```

```python
# Tick 설정
plt.plot(x, y)
plt.xticks(rotation=90)
plt.yticks(rotation=45, fontsize=15)
plt.show()
```

```python
# Legend(범례) 설정
plt.plot(x, y)
plt.plot(y, x)
plt.legend(['(x, y)', '(y, x)'])
plt.show()
```

```python
# Legend 설정 2
plt.plot(x, y, label='(x, y)')
plt.plot(y, x, label='(y, x)')
plt.legend()
plt.show()
```

## 2-4. Grid

```python
# Grid 그리기
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
# grid 속성 변경하기
plt.plot(x, y)
plt.grid(c='r', ls=':', lw=2)
plt.show()
```

## 2-5. Subplot
* 하나의 figure에 여러개 그래프 그리기 

```python
# 1. subplot(row, column, index)
plt.subplot(1, 2, 1)
plt.plot(x, y)

plt.subplot(1, 2, 2)
plt.plot(y, x)

plt.show()
```

```python
# (1, 2, 1) -> 121
plt.subplot(211)
plt.plot(x, y)
plt.title('(x, y)')

plt.subplot(212)
plt.plot(y, x)
plt.title('(y, x)')

plt.suptitle('Title', size=20)
plt.tight_layout()  # 자동 레이아웃 설정
plt.show()
```

```python
# 2. add_subplot()
fig = plt.figure()
axes1 = fig.add_subplot(2, 1, 1)  # row, col, ind
axes2 = fig.add_subplot(2, 1, 2)

axes1.plot(x, y)
axes1.set_title('(x, y)')

axes2.plot(y, x)
axes2.set_title('(y, x)')

plt.tight_layout()
plt.show()
```

```python
# 3. subplots()
fig, axes = plt.subplots(2, 1)  # row, col

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
plt.show()
```
